```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.array([1,3,5,7,9])
y = np.array([2,6,10,14,16])
```

```python
plt.plot(x,y,".")
plt.xlim(0,10)
plt.ylim(0,18)
```

```python
plt.scatter(x,y)
```

```python
x2 = np.array([2,4,6,8,10])
y2 = np.array([2,6,10,14,16])
```

```python
plt.plot(x,y,"b.")
plt.plot(x2,y2,"r.")
plt.xlim(0,12)
plt.ylim(0,18)
```

```python
plt.scatter(x,y)
plt.scatter(x2,y2)
```

```python
sns.pointplot(x,y, join=False, color="#FF0000")
sns.pointplot(x2,y2, join=False)
```

```python
x3 = np.array(["Mon","Tue","Wed","Thu","Fri"])
y3 = np.array([2,6,10,14,16])
```

```python
sns.pointplot(x,y, join=False, color="#FF0000")
sns.pointplot(x3,y3, join=False)
```

```python
plt.plot(x,y,".")
plt.plot(x3,y3,".")
```
