```python
data = np.loadtxt("samplenetwork3.csv", delimiter=",", dtype="int32")
print(data)
```

```python
link = data[data[:,0]>1000,:]
print(link)
```

```python
node = data[data[:,0]<1000,:]
print(node)
```

```python
allnode = node[:,0]
print(allnode)
```

```python
subjnode = 503
index = np.where(node[:,0]==subjnode)
print(index)
chknode = node[index].ravel()
print(chknode)
```

```python
nextlink = chknode[1:]
print(nextlink)
```

```python
index = np.where(link[:,0]==nextlink[0])
print(index)
chklink = link[index].ravel()
print(chklink)
```
