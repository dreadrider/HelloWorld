# 기계학습이란 무엇인가?
기계가 학습을 하기 위해서는 무엇이 필요할까? 또한 그것은 기계를 제어하는 것과 어떻게 다를까?
![산업용로봇](https://www.keyence.co.jp/ss/visionsystem/fa-robot/industrial_robot/mechanism/images/img02.jpg)
단순하게 기계를 제어하기만 한다면 위 그림처럼 각 관절의 각도와 이동속도를 입력하면 그대로 움직일 것이다.
그런데 정확하게 어느 위치의 물체를 다른 곳으로 이동시키기 위해 기계를 제어한다면 그 이동이 가능하도록 관절의 각도와 이동속도를 입력해 놓으면 가능할 것이다. 만약 컨베어벨트 위의 물체가 항상 같은 위치에 있지 않고 그 위치가 조금씩 오차를 갖고 랜덤하게 변한다면 센서를 달아 물체의 위치를 파악하고 각 위치에 대한 움직임을 미리 프로그래밍해 놓으면 제어가 가능할 것이다.
>여기서 *프로그래밍* 이란 어떤 위치에서 어떤 위치까지 움직이기 위한 관절의 각도와 이동속도를 계산하는 것을 말한다.

다시말해서 기계를 제어하는 것만으로도 얼마든지 원하는 결과를 얻을 수 있고 자동반복시킬 수 있다. 이러한 기계제어 덕분에 산업화가 빠르게 발전했고 제품이 다량으로 생산될 수 있었다. 그러나 위에서 쉽게 기계를 제어할 수 있는 것처럼 말했지만 실제로는 그렇게 쉬운 작업은 아니다. 칠판에 분필로 선을 긋는 작업을 하는데도 정확하게 제어를 하기 위해서는 어려운 수학식을 사용한다. 회전군인 리군등을 이용한다고 한다.

그런데 이족보행 로보트를 봐 보자.<br>
걸을때마다 지면의 상태를 확인해가면서 각 관절의 상태를 조절하는 방법으로 제어할 수도 있겠지만 공장에서 컨베어벨트 위를 이동하는 물체에 비해 매우 많은 변수가 존재한다는 것을 알 수 있다. 지면이 울퉁불퉁한지 아니면 경사가 졌는지 낭떨어지가 있는 것은 아닌지 등등을 알아봐야 하고 바람이 부는지 비가 오는지 눈이 오는지 등의 날씨에 관해서도 알아야 한다. 게다가 누가 발로 차면 되도록이면 넘어지지 않고 밸런스를 잡는 움직임도 필요하다.
![이족보행](http://blogs.c.yimg.jp/res/blog-f0-36/tyumo19601961/folder/1563605/13/63984713/img_6_m?1483666614)

이러한 모든 조건과 상황을 센서만을 가지고 제어하는 것은 매우 어려운 일이고 거의 불가능한 일이다. 따라서 어떤 상황에서는 이러한 제어를 하고 또 다른 상황에서는 다른 제어를 하도록 프로그래밍해 놓은 상태에서 많은 다양한 상황을 경험하게 하는 것이다. 각 상황에 맞는 정확한 대응 방법을 프로그래밍해 놓는 것이 아니라 기본적으로 어떤 범위내에서 어떤 움직임을 하도록 엉성하게 만들어 놓고 얼마든지 실수가 일어날 수 있으며 오히려 그 실수를 장려하는 상황을 만들어 놓는다. 그러는 과정에서 실수에 대한 원인을 찾아가고 조금씩 엉성했던 프로그래밍안의 파라메터를 조절해 나가도록 만들 수 있다. 이것이 바로 *기계학습* 이란 것이다.

# 어떻게 학습을 시킬 것인가?
<p>기본적으로 데이터들이 있고 어떤 상황에서 어떤 움직임을 했을 때의 결과가 있다. 여기서 학습이란 것은 어떤 상황에서 될 수록 같은 움직임을 하고 그 결과가 좋도록 만든는 과정이다. 따라서 위에서 언급한 로봇을 걷게 만드는 학습은 여러 상황에 맞는 각 관절의 움직임을 결과에 맞춰 조절해가는 작업이다. 입력값은 여러 상황 데이터들이 있을 것이고 그 위에 다음 동작을 위해 각 관절을 움직이는 값들이 될 것이다. 결과값은 그 동작을 맞춘 상태의 상황 데이터와 로봇의 위치와 밸런스에 관한 데이터 그러한 상황이 실수인지 아닌지 판단한 라벨 등이 있을 수 있다. 이러한 데이터를 이용해서 다양한 상황에서의 결과를 가지고 파라메터를 조절하는 학습을 여러번의 반복으로 진행시키다보면 기계가 걷는 동작을 배우게 된다는 원리이다.</p>
<p>그런데 무엇을 가지고 학습을 시키면 되는지는 알았지만 어떻게 학습을 시킬 것인가에 대해서는 아직 두리뭉실한 뜬 구름 잡는 이야기이다. 따라서 너무 어려운 학습과정을 생각하기 보다는 기초적인 학습에 대해 생각해보고 나중에 이에 관해 다시 고찰해보는 것이 좋을 것이다. 그러면 기초적인 학습이란 무엇인가? 크게 두가지로 나눌 수 있는데 하나는 교사가 있는 학습법과 다른 하나는 교사가 없는 학습법이다. 또한 이와 달리 두가지로 나눌 수도 있다. 하나는 예측(Regressioni)이고 또 하나는 분류(Classification)이다.</p>
- Regression : 어떤 입력된 값(보통은 숫자이나 문자열이나 화상, 영상 데이터도 될 수 있다.)에 대해 결과를 예측하는 것이다.<br>
*예를 들면 2020년의 GDP는 얼마인가? 라든지 키 170cm의 평균 몸무게는 몇 kg인가? 등등이 될 수 있다.*
- Classification : 어떤 입력된 값(보통은 화상, 영상 데이터이고 문자열의 단어나 조사 등이 될 수도 있다.)에 대해 분류된 결과를 준다.<br>
*예를 들면 어떤 손 글씨 화상 데이터가 숫자 0부터 9중에 어느 숫자로 분류되는지 알아내는 것이다. 혹은 "나는 오늘 아침에 빵과 계란을..." 이라는 문장을 주면 다음에 나올 단어를 가지고 있는 단어장에서 분류해서 찾아내는 것이다.*

교사가 있는 학습법과 없는 학습법은 말 그대로 선생님이 있는가 없는가 인데 선생님이 결과를 보고 맞았는지 틀렸는지를 채점하는 경우가 교사가 있는 학습법이 되고 결과가 선생님의 도움 없이 도출되거나 분류될 때 교사가 없는 학습법이 된다.

먼저 교사 있는 학습법 그리고 Regression에 대해서 알아보자. 보통 머신 러닝에 대한 설명은 뉴럴 네트워크 즉 뇌에 대한 뉴런의 학습법을 먼저 설명하기도 하나 왜 뇌는 그런 학습법을 가지고 있고 머신 러닝을 연구하는 사람들은 그 학습법을 이용하게 되었는지 알아가는게 더욱 중요하다.<br>
