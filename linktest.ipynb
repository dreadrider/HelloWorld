  1. 호텔
     - 우리집 근처에 있는 저렴하지만 깨끗한 호텔을 살펴 보았다.
     - 우선 우리집의 주소는 오사카후 오사카시 탠노지쿠 오바세쵸 이다.
     - 일본은 경기도의 도와 같은 개념이 4개 있는데 하나는 동경도(東京都)<br>
       그리고 북해도(北海道) 여기까지는 우리나라처럼 도이다.<br>
       그런데 오사카후(大阪府) 와 교토후(京都府)는 도를 후(우리나라 한자발음은 부)라고 부른다.<br>
       마지막으로 나머지는 전부 켄(우리나라 한자발음은 현)이라고 부른다.<br>
       예를 들면 내가 다니는 학교는 시가켄(滋賀県)이다.<br>
       우리나라에서도 도 안에 여러 시가 있듯이 일본에도 토도후켄 밑에 시가 있다.<br>
       따라서 오사카후 밑에 오사카시가 있지만 다른 시도 있다. 예를 들면 타카츠키시가 있다.<br>
       동경도 밑에는 동경시가 없다. 그런데 서울특별시처럼 사람들이 모여 있는 곳이 있는데<br>
       거긴 바로 구로 내려간다. 그렇게 동경도에서 구로 바로 내려가는 곳이 23개 있어서 23구라 한다.<br>
       그리고 일본은 몇개의 도를 묶어서 지방이라고도 하는데 동경도가 포함된 지역을 관동지방이라 한다.<br>
       오사카후가 포함된 지역을 킨키지방이라고 하는데 관동지방과 비교할때는 관서지방이라고 부른다.<br>
       따라서 관서지방을 여행하는 것을 보통 오사카 여행이라고 한다.<br>
       그러면 보통 오사카시랑 교토시 그리고 코베시 혹은 나라시의 여행을 말한다.<br>
       교토시는 관광시설이 많고 오사카시는 먹거리가 유명하고 코베는 페션의 거리인데,<br>
       오사카에 거점을 두고 열차를 타면 편도 한시간 정도의 거리에 있다.<br>

     - 여덟번째는 [호텔 토라스티](https://goo.gl/v90isc) 우리집에서는 좀 떨어져 있지만 친구가 묵고 괜찮았다고 한 곳<br>

     - 우선 [에어비앤비](https://goo.gl/soiFXd)는 하숙과 비슷한 느낌인데 주인이 없고 방만 빌려준다는거 또한 기한이 매우 짧아도 된다는 것이다.<br>

       4명가족(애들 2명)이면 대충 1박에 1만앤 이하인 곳을 찾으면 괜찮다고 본다. 더 비싼곳은 더 많은 사람이 들어갈 수 있다.<br>
